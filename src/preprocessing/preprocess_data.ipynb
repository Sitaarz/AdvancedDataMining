```python
import re
import string

import nltk
import pandas as pd

nltk.download('punkt_tab')

from bs4 import BeautifulSoup

def clean_text(text: str) -> str:
    text = BeautifulSoup(text, 'html.parser').get_text() # parsing html
    text = text.lower() # to lower
    text = text.translate(str.maketrans('','', string.punctuation)) # punktuation
    return text

def remove_not_language_sigs(word: str)-> str:
    word = re.sub(r'[^a-z ]', "", word)
    word = re.sub(r'\s+', " ", word)
    return word
```

```python
df = pd.read_csv("../../data/filtered_data_10000_rows_per_genre.csv")
print(df.columns)

df.head()
```

```python
df = df.map(clean_text)
df.head()
```

```python
latest_3_columns = df.iloc[:, -3:].applymap(remove_not_language_sigs)

df = pd.concat([df.iloc[:, [0]], latest_3_columns], axis=1)
```

```python
df.head()
```

```python
df.to_csv("../../data/preprocessed_data.csv", index=False)
```
