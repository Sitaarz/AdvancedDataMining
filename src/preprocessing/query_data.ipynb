```python
import sqlite3

import pandas as pd


def get_df_with_filtered_data(dbFilePath: str):
    with sqlite3.connect(dbFilePath) as conn:
        query = """
                SELECT TRIM(imdbId) as imdbId,
                       TRIM(title) as title,
                       TRIM(genre) as genre,
                       TRIM(director) as director,
                       TRIM(plot) as plot
                FROM movies
                WHERE
                    imdbId IS NOT NULL
                    AND title IS NOT NULL
                    AND title NOT LIKE '%N/A%'
                    AND genre IS NOT NULL
                    AND genre NOT LIKE '%N/A%'
                    AND director IS NOT NULL
                    AND director NOT LIKE '%N/A%'
                    AND plot IS NOT NULL
                    AND plot NOT LIKE '%N/A%'
                    AND LENGTH(plot) > 200
                GROUP BY title
                """
        return pd.read_sql_query(query, conn, ['imdbId'])


df = get_df_with_filtered_data("../../data/movies.db")
df.head()
# df.describe()
# df.info()

df['genre'].value_counts()
```

```python
most_frequent = df['genre'].value_counts().head(3).index
print(most_frequent)
```

```python
top_3 = df[df['genre'].isin(most_frequent)]
data_to_save_10000_rows = top_3.groupby('genre', group_keys = False).head(10_000)
# data_to_save_6000_rows.describe()
# data_to_save_6000_rows.head()
data_to_save_10000_rows
```

```python
data_to_save_10000_rows.to_csv("../../data/filtered_data_10000_rows_per_genre.csv", index = False)
```
